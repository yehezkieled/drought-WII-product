Normal Historical Burn Analysis Method functions

```python
def historical_claim(accumulation_list):
#function that returns historical claim using each accumulation value per year as a trigger
    hist_claim = []
    for i in range(0, len(accumulation_list)):
        #using each accumulation value as a trigger
        trigger = accumulation_list[i]
        if trigger != 0:
            #formula to get claim per trigger
            acc_sum = sum(accumulation_list[0:i])
            claim = (i*trigger - acc_sum)/trigger
            hist_claim.append(claim)
        else:
            claim = 0
            hist_claim.append(claim)
    return hist_claim
    
```

```python
def getting_trigger(hist_prem, cover):
#function that returns trigger that satisfies the pure premium per phase
    #weight of pure premium per phase 
    pure_prem_phase = pure_prem*cover
    
    #considering if the first accumulation is 0
    if hist_prem['accumulation'][0] == 0:
        fin_trigger = sum(hist_prem['accumulation'][0:2])/(2-len(year)*pure_prem_phase)
    else:
        for i in range(0, len(hist_prem)):
            if hist_prem['hist_prem'][i] > pure_prem_phase:
                fin_trigger = sum(hist_prem['accumulation'][0:i])/(i-len(year)*pure_prem_phase)
                break
    
    return fin_trigger
```

Tiered Historical Burn Analysis Method functions

```python
def adjusted_accumulation(accumulation_list):
#function that returns a list of adjusted accumulation (max and min trigger) to get the first trigger
    trigger_1_list = list(map(lambda x: max_trigger if x > max_trigger else x, accumulation_list))
    trigger_1_list = list(map(lambda x: min_trigger if x < min_trigger else x, trigger_1_list))

    return trigger_1_list
```

```python
def getting_trigger_2(accumulation_list):
#function that returns the second tiered trigger list
    trigger_2_list = []
    for i in accumulation_list:
        trigger_2 = (1-(tiered_1)*i)/(-tiered_1+tiered_2)
        #considering if the second trigger is more than the first trigger
        if trigger_2 > i:
            trigger_2_list.append(i*decrease_trigger)
        else:
            trigger_2_list.append(trigger_2)
    
    return trigger_2_list
```

```python
def historical_claim_tiered(trigger1, trigger2, accumulation_list):
#function that returns the historical claim list for tiered historical burn analysis method
    hist_acc_tiered_list = []
    for j in range(len(trigger1)):
        hist_acc_tiered = 0
        for i in accumulation_list:
            #considering scenarios on where the rainfall accumulation is
            if i > tiered_2_trigger:  
                if i < trigger1[j] and i > trigger2[j]:
                    hist_acc_tiered += (-tiered_1)*i + tiered_1*trigger1[j]
                elif 0 < i and i < trigger2[j]:
                    hist_acc_tiered += (-tiered_2)*i + 1
            else:
                if i < trigger1[j] and i > trigger2[j]:
                    hist_acc_tiered += (trigger1[j]-i)*tiered_1
                elif 0 <= i and i < trigger2[j]:
                    hist_acc_tiered += (trigger1[j] - trigger2[j])*tiered_1 + (trigger2[j] - i)*tiered_2
        hist_acc_tiered_list.append(hist_acc_tiered)
    
    return hist_acc_tiered_list
```

```python
def getting_SI(trigger_1_list, trigger_2_list):
#function that returns the sum insured list 
    SI = []
    for i in range(len(trigger_1_list)):
        if trigger_1_list[i] > tiered_2_trigger:
            SI.append(1)
        else:
            TC = (tiered_2-tiered_1)*trigger_2_list[i] + trigger_1_list[i]*tiered_1
            SI.append(TC)
    
    return SI
```

```python
def hist_claim_per_year(hist_acc_table, trigger1, trigger2, cover):
#function that returns historical claim yearly, total historical claim, and pure premium
    hist_claim_acc = 0              #total historical claim
    hist_claim = []                 #historical claim yearly list
    for i in range(len(hist_acc_table['accumulation'])):
        accumulation = hist_acc_table['accumulation'][i]
        if cover == 0:
            claim = 0
            hist_claim.append(claim)
        else:
            if accumulation < trigger1:
                if accumulation > trigger2:
                    claim = (trigger1 - accumulation)*tiered_1
                    hist_claim_acc += claim
                    hist_claim.append(claim)
                elif accumulation <= trigger2:
                    claim = (trigger1 - trigger2)*tiered_1 + (trigger2-accumulation)*tiered_2
                    hist_claim_acc += claim
                    hist_claim.append(claim)
            else:
                claim = 0
                hist_claim_acc += claim
                hist_claim.append(claim)
    hist_claim.append(hist_claim_acc)
    hist_claim.append(hist_claim_acc/len(year))         #pure premium
    return(hist_claim)
```

```python
def trigger1_check(table_tiered, adjusted_trigger1, adjusted_trigger2, cover):
#function that check whether the adjusted trigger 1 and adjusted trigger 2 pure premium is below the required pure premium
    r = pure_prem*cover             #pure premium per phase 
    hist_claim_list = hist_claim_per_year(table_tiered, adjusted_trigger1, adjusted_trigger2, cover)        #taking the pure premium out of the list by indexing -1
    if round(hist_claim_list[-1], 3) > round(r,3): 
        adjusted_trigger1 = tiered_2_trigger
        adjusted_trigger2 = adjusted_trigger1*decrease_trigger
        return(adjusted_trigger1, adjusted_trigger2)
    else:
        adjusted_trigger1 = adjusted_trigger1
        adjusted_trigger2 = adjusted_trigger2
        return(adjusted_trigger1, adjusted_trigger2)
        
```

```python
def adjusted_trigger_tiered_FC(table_tiered, temp_trigger2_upper, temp_trigger2_lower, cover, i):
#function that returns adjusted trigger 1 and its trigger 2 using FULL COVER formula
    acc_upper = []
    acc_lower = []
    hist_claim = pure_prem*cover*len(year)

    for k in range(i):
        if table_tiered['accumulation'][k] < temp_trigger2_lower:
            acc_lower.append(table_tiered['accumulation'][k])
        elif table_tiered['accumulation'][k] >= temp_trigger2_upper:
            acc_upper.append(table_tiered['accumulation'][k])
    adjusted_trigger1 = (hist_claim + tiered_2*sum(acc_lower) + tiered_1*sum(acc_upper) - len(acc_lower))/(tiered_1*len(acc_upper))
    adjusted_trigger2 = (1-tiered_1*adjusted_trigger1)/(-tiered_1+tiered_2)
    #the triggers are getting checked
    adjusted_trigger2 = trigger1_check(table_tiered, adjusted_trigger1, adjusted_trigger2, cover)[1]
    adjusted_trigger1 = trigger1_check(table_tiered, adjusted_trigger1, adjusted_trigger2, cover)[0]
    return(adjusted_trigger1, adjusted_trigger2)
```

```python
def adjusted_trigger_tiered_TC(table_tiered, temp_trigger2_upper, temp_trigger2_lower, cover, i):
#function that returns adjusted trigger 1 and trigger 2 using the TOTAL COVER formula
    acc_upper = []
    acc_lower = []
    hist_claim = pure_prem*cover*len(year)
    for k in range(i):
        if table_tiered['accumulation'][k] <= temp_trigger2_lower:
            acc_lower.append(table_tiered['accumulation'][k])
        elif table_tiered['accumulation'][k] >= temp_trigger2_upper:
            acc_upper.append(table_tiered['accumulation'][k])
    sum_rf = tiered_2*sum(acc_lower) + tiered_1*sum(acc_upper)
    denominator = len(acc_lower)*decrease_trigger*(tiered_2-tiered_1) + tiered_1*(len(acc_lower) + len(acc_upper))
    adjusted_trigger1 = (hist_claim + sum_rf)/denominator
    adjusted_trigger2 = adjusted_trigger1*decrease_trigger
    #the triggers are getting checked
    adjusted_trigger2 = trigger1_check(table_tiered, adjusted_trigger1, adjusted_trigger2, cover)[1]
    adjusted_trigger1 = trigger1_check(table_tiered, adjusted_trigger1, adjusted_trigger2, cover)[0]
    return(adjusted_trigger1, adjusted_trigger2)
```

```python
def adjusted_trigger_tiered(table_tiered, cover):
#function that taking account all of the scenarios to get adjusted triggers
    r = pure_prem*cover

    for i in range(len(table_tiered['hist_prem'])):
        if table_tiered['hist_prem'][i] >= r:
            benchmark_trigger = table_tiered['trigger1'][i]
            in_between_trigger2 = []
            temp_pure_prem = []
            if i == 0:
                adjusted_trigger1 = table_tiered['trigger1'][i]
                adjusted_trigger2 = table_tiered['trigger2'][i]
                return(adjusted_trigger1, adjusted_trigger2)
                break
            
            if benchmark_trigger <= tiered_2_trigger:
                for j in range(i):
                    if (table_tiered['accumulation'][j]>table_tiered['trigger2'][i - 1] and
                    table_tiered['accumulation'][j]<table_tiered['trigger2'][i]):
                        in_between_trigger2.append(table_tiered['accumulation'][j])
                    
                if len(in_between_trigger2) == 0:
                    temp_trigger2_upper = table_tiered['trigger2'][i]
                    temp_trigger2_lower = table_tiered['trigger2'][i-1]
                    return(adjusted_trigger_tiered_TC(table_tiered, temp_trigger2_upper, temp_trigger2_lower, cover, i))
                    break
                else:
                    in_between_trigger2.append(table_tiered['trigger2'][i])
                    for k in in_between_trigger2:
                        temp_trigger2 = k
                        temp_trigger1 = temp_trigger2/decrease_trigger
                        hist_acc_tiered = 0
                        for l in table_tiered['accumulation']:
                            if l > tiered_2_trigger:
                                if l < temp_trigger1 and l > temp_trigger2:
                                    hist_acc_tiered += (-tiered_1)*l + tiered_1*temp_trigger1
                                elif 0 <= l and l < temp_trigger2:
                                    hist_acc_tiered += (-tiered_2)*l + 1
                            else:
                                if l < temp_trigger1 and l >= temp_trigger2:
                                    hist_acc_tiered += (temp_trigger1-l)*tiered_1
                                elif 0 <= l and l < temp_trigger2:
                                    hist_acc_tiered += (temp_trigger1 - temp_trigger2)*tiered_1 + (temp_trigger2 - l)*tiered_2
                        temp_pure_prem.append(hist_acc_tiered/len(year))
                    
                    for x in range(len(temp_pure_prem)):
                        if temp_pure_prem[x]>r:
                            temp_trigger2_upper = in_between_trigger2[x]
                            temp_trigger2_lower = table_tiered['trigger2'][i-1] if x == 0 else in_between_trigger2[x-1]
                            
                            return(adjusted_trigger_tiered_TC(table_tiered, temp_trigger2_upper, temp_trigger2_lower, cover, i))
                            break
            
            elif benchmark_trigger > tiered_2_trigger:
                if table_tiered['trigger2'][i] > table_tiered['trigger2'][i-1]:
                    upper_bound = table_tiered['trigger2'][i]
                    lower_bound = table_tiered['trigger2'][i-1]
                else:
                    upper_bound = table_tiered['trigger2'][i-1]
                    lower_bound = table_tiered['trigger2'][i]

                for j in range(i):
                    if (table_tiered['accumulation'][j]>lower_bound and
                    table_tiered['accumulation'][j]<upper_bound):
                        in_between_trigger2.append(table_tiered['accumulation'][j])
                
                if (table_tiered['trigger1'][i] > tiered_2_trigger and table_tiered['trigger1'][i-1] < tiered_2_trigger and len(in_between_trigger2) ==0):
                    temp_trigger1 = tiered_2_trigger
                    temp_trigger2 = 0.4*temp_trigger1
                    hist_acc_tiered = 0
                    for l in table_tiered['accumulation']:
                        if l < temp_trigger1 and l>= temp_trigger2:
                            hist_acc_tiered += (temp_trigger1-1)*tiered_1
                        elif 0 <= l and l < temp_trigger2:
                            hist_acc_tiered += (temp_trigger1 - temp_trigger2)*tiered_1 + (temp_trigger2 - l)*tiered_2
                    
                    if hist_acc_tiered >= r:
                        temp_trigger2_upper = temp_trigger2
                        temp_trigger2_lower = table_tiered['trigger2'][i-1]
                        return(adjusted_trigger_tiered_TC(table_tiered, temp_trigger2_upper, temp_trigger2_lower, cover, i))
                        break
                    else:
                        temp_trigger2_upper = table_tiered['trigger2'][i]
                        temp_trigger2_lower = temp_trigger2
                        return(adjusted_trigger_tiered_FC(table_tiered, temp_trigger2_upper, temp_trigger2_lower, cover, i))
                        break
                elif (table_tiered['trigger1'][i] > tiered_2_trigger and table_tiered['trigger1'][i-1] < tiered_2_trigger and len(in_between_trigger2) !=0):
                    in_between_trigger2.append(decrease_trigger*tiered_2_trigger)
                    in_between_trigger2.append(table_tiered['trigger2'][i])
                    in_between_trigger2.sort()

                    for k in in_between_trigger2:
                        temp_trigger2 = k
                        temp_trigger1 = temp_trigger2/decrease_trigger
                        hist_acc_tiered = 0
                        for l in table_tiered['accumulation']:
                            if l > tiered_2_trigger:
                                if l < temp_trigger1 and l > temp_trigger2:
                                    hist_acc_tiered += (-tiered_1)*l + tiered_1*temp_trigger1
                                elif 0 <= l and l < temp_trigger2:
                                    hist_acc_tiered += (-tiered_2)*l + 1
                            else:
                                if l < temp_trigger1 and l >= temp_trigger2:
                                    hist_acc_tiered += (temp_trigger1-l)*tiered_1
                                elif 0 <= l and l < temp_trigger2:
                                    hist_acc_tiered += (temp_trigger1 - temp_trigger2)*tiered_1 + (temp_trigger2 - l)*tiered_2
                        temp_pure_prem.append(hist_acc_tiered/len(year))

                    for x in range(len(temp_pure_prem)):
                        if temp_pure_prem[x]>r:
                            temp_trigger2_upper = in_between_trigger2[x]
                            temp_trigger2_lower = table_tiered['trigger2'][i-1] if len(temp_pure_prem) == 1 or x == 0 else in_between_trigger2[x-1]
                            
                            return(adjusted_trigger_tiered_TC(table_tiered, temp_trigger2_upper, temp_trigger2_lower, cover, i))
                            break
                elif len(in_between_trigger2) != 0:
                    in_between_trigger2.append(table_tiered['trigger2'][i])
                    # in_between_trigger2.sort()
                    for k in in_between_trigger2:
                        temp_trigger2 = k
                        temp_trigger1 = (temp_trigger2*(-tiered_1+tiered_2) - 1)/(-tiered_1)
                        hist_acc_tiered = 0
                        for l in table_tiered['accumulation']:
                            if l > tiered_2_trigger:
                                if l < temp_trigger1 and l > temp_trigger2:
                                    hist_acc_tiered += (-tiered_1)*l + tiered_1*temp_trigger1
                                elif 0 <= l and l < temp_trigger2:
                                    hist_acc_tiered += (-tiered_2)*l + 1
                                elif l > temp_trigger1:
                                    hist_acc_tiered += 0
                            else:
                                if l < temp_trigger1 and l >= temp_trigger2:
                                    hist_acc_tiered += (temp_trigger1-l)*tiered_1
                                elif 0 <= l and l < temp_trigger2:
                                    hist_acc_tiered += (temp_trigger1 - temp_trigger2)*tiered_1 + (temp_trigger2 - l)*tiered_2
                        temp_pure_prem.append(hist_acc_tiered/len(year))

                    for x in range(len(temp_pure_prem)):
                        if temp_pure_prem[x]>r:
                            temp_trigger2_upper = in_between_trigger2[x]
                            temp_trigger2_lower = table_tiered['trigger2'][i-1] if len(temp_pure_prem) == 1 or x == 0 else in_between_trigger2[x-1]
                            
                            return(adjusted_trigger_tiered_FC(table_tiered, temp_trigger2_upper, temp_trigger2_lower, cover, i))
                            break
                else:
                    temp_trigger2_upper = table_tiered['trigger2'][i]
                    temp_trigger2_lower = table_tiered['trigger2'][i-1]

                    return(adjusted_trigger_tiered_FC(table_tiered, temp_trigger2_upper, temp_trigger2_lower, cover, i))
                    break
            break
    return(table_tiered['trigger1'][len(year)-1], table_tiered['trigger2'][len(year)-1])
```

```python
def flooring(trigger, flooring_digits):
#function that returns the nearest lower interger value of a trigger
    if flooring_digits == "none":
        floored_number = trigger
    else:
        factor = 10**flooring_digits
        floored_number = math.floor(trigger*factor)/factor
    return floored_number
```
